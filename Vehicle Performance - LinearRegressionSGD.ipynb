## 5.Label encoding with Categorical variables

```python
data_with_dummies = pd.get_dummies(data_cleaned, drop_first=True)
data_with_dummies.head()
```

## 6. Standardise data

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
x = data_with_dummies.drop(["Kilometer_per_liter","log_mileage"], axis =1)
```

```python
scaler = MinMaxScaler()
scaler.fit(x)
```

```python
X = scaler.fit_transform(x)
```

```python
y = data_with_dummies.log_mileage
```

### split data into train and test

```python
from sklearn.model_selection import train_test_split
#let us now split the 
X_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.20, random_state= 10)

print("X_train", X_train.shape)
print("X_test", x_test.shape)
print("y_train", y_train.shape)
print("y_test", y_test.shape)
```

## 7.Linear Regression

1.build the model

```python
from sklearn.linear_model import LinearRegression

OLS_model = LinearRegression()

OLS_model.fit(X_train, y_train)
```

2.predict the values

```python
y_pred_OLS = OLS_model.predict(x_test)
```

```python
plt.scatter(y_test, y_pred_OLS)
plt.xlabel("Targets", size =18)
plt.ylabel("Predictions", size = 18)
plt.show()
```

3.compute accuracy measures

r_squared_OLS = OLS_model.score(X_train, y_train)
#R-squared is a statistical measure that represents the proportion of the variance for a dependent
#variable that's explained by an independent variable or variables in a regression model

n=313
p =9

adj_r_squared_OLS = 1 - (1-r_squared_OLS)*(n-1)/(n-p-1)
#Adjusted R-squared is a modified version of R-squared that has been adjusted for the number of
#predictors in the model

from sklearn.metrics import mean_squared_error
from math import sqrt

rmse_OLS = sqrt(mean_squared_error(y_test, y_pred_OLS))
#The root-mean-square deviation or root-mean-square error is a frequently used measure of the
#differences between values predicted by a model or an estimator and the values observed


4. tabulate the results

```python
cols = ["Model", "R_squared", "Adj R-squared", "RMSE"]

result_tabulation = pd.DataFrame(columns=cols)

OLS_metrics = pd.Series({"Model": "OLS", "R_squared": r_squared_OLS, 
                        "Adj R-squared": adj_r_squared_OLS, "RMSE" : rmse_OLS})

result_tabulation = result_tabulation.append(OLS_metrics, ignore_index=True)

result_tabulation
```

## 7.2 Linear regression using SGD

1.build the model

```python
from sklearn.linear_model import SGDRegressor

SGD_model  = SGDRegressor(loss = "squared_loss", alpha = 0.1, max_iter=1000 )
#alpha:float, default=0.0001
#Constant that multiplies the regularization term. The higher the value, the stronger the
#regularization
#max_iter:int, default=1000
#The maximum number of passes over the training data (aka epochs)

SGD_model.fit(X_train, y_train)
```

2. predict hte values

```python
y_pred_SGD = SGD_model.predict(x_test)
```

```python

plt.scatter(y_test, y_pred_SGD)
plt.xlabel("Targets", size =18)
plt.ylabel("Predictions", size = 18)
plt.show()
```

3.compute accuracy measures

```python
r_squared_SGD = SGD_model.score(X_train, y_train)
#R-squared is a statistical measure that represents the proportion of the variance for a dependent
#variable that's explained by an independent variable or variables in a regression model

n=313
p =9

adj_r_squared_SGD = 1 - (1-r_squared_SGD)*(n-1)/(n-p-1)
#Adjusted R-squared is a modified version of R-squared that has been adjusted for the number of
#predictors in the model

from sklearn.metrics import mean_squared_error
from math import sqrt

rmse_SGD = sqrt(mean_squared_error(y_test, y_pred_SGD))
#The root-mean-square deviation or root-mean-square error is a frequently used measure of the
#differences between values predicted by a model or an estimator and the values observed
```

4 tabulate the results

```python


SGD_metrics = pd.Series({"Model": "SGD", "R_squared": r_squared_SGD, 
                        "Adj R-squared": adj_r_squared_SGD, "RMSE" : rmse_SGD})

result_tabulation = result_tabulation.append(SGD_metrics, ignore_index=True)

result_tabulation
```
